```python
#   Write a Python Program to Check if a Number is Positive, Negative or Zero?


a=int(input("Please enter a number to find whether it is Positive, Negative or Zero?:"))
if (a==0):
    print("The given number is zero")
elif (a<0):
    print("The given number is negative")
else: 
    print("The given number is positive")
```

```python
#   Write a Python Program to Check if a Number is Odd or Even?


a=int(input("Please enter a number to find whether it is odd or even:"))
if (a%2==0):
    print("The given number is Even")
else: 
    print("The given number is Odd")
```

```python
#   Write a Python Program to Check Leap Year?

a=int(input("Please enter a year to find whether it is a leap year or not:"))
if (a%4==0 and a%400==0):
    print("The given year is leap year")
else: 
    print("The given number is not leap year")
```

```python
#   Write a Python Program to Check Prime Number?

a=int(input("Please enter a number to find whether it is a prime number or not:"))
if (a%2!=0 or a==2):
    print("The given number is prime number")
elif a==1:
    print("The given number is prime number")       
else: 
    print("The given number is not prime number")
```

```python
#         Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?

for a in range (1,10000):
    if (a%2!=0 or a==2):
        print(a," is prime number")
    elif a==1:
        print(a," is prime number")
    else: 
        continue
```
